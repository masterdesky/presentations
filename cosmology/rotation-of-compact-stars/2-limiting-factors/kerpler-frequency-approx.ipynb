## Approximation of $\Omega_{K}$

$$
\Omega_{K}
\approx
0.65 \cdot \sqrt{\frac{M}{R^{3}}}
$$

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# M / M_{Sun}
M = np.linspace(1, 3, 1000)
# Radius [km]
R = np.linspace(4, 11, 1000)

# Omega_{K}
M, R = np.meshgrid(M, R)
Omega_K = 0.65 * np.sqrt(M / R**3)
```

```python
L = 240 * np.sqrt(M / R**3)
```

```python
fig, ax = plt.subplots(figsize=(14, 12), facecolor='white')
c = ax.contourf(M, R, Omega_K, 20)
ax.set_xlabel('$M / M_{\odot}$', fontsize=42)
ax.set_ylabel('Radius [km]', fontsize=42)
ax.set_xticks(np.arange(np.min(M), np.max(M)+0.5, 0.5))
ax.set_xticks(np.arange(np.min(M), np.max(M)+0.1, 0.1), minor=True)
ax.tick_params(axis='both', which='major', labelsize=28, pad=10,
               length=20, width=2)
ax.tick_params(axis='both', which='minor', labelsize=28, pad=10,
               length=10, width=2)

for x in ax.spines.values():
    x.set_linewidth(2)

cbar = plt.colorbar(c, ax=ax)
cbar.ax.tick_params(labelsize=22)
cbar.ax.set_ylabel('$\Omega_{K}$ [$1 / 10\,ms$]', fontsize=36)

plt.savefig(os.path.join('images', 'ns-m-radius-Omega_k.pdf'),
            bbox_inches='tight')

plt.show()
```
